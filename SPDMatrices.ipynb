```python
from geomstats.geometry.spd_matrices import SPDMatrices
from geomstats.geometry.riemannian_metric import RiemannianMetric
from geomstats.geometry.spd_matrices import SPDMetricBuresWasserstein

import matplotlib.pyplot as plt

import geomstats.backend as gs
```

```python

space = SPDMatrices(5)
m = space.metric
samples = space.random_point(5)

d = m.geodesic(samples[0] , samples[1])
g = d(gs.linspace(0., 1., 20))

for m in g:
  plt.matshow(m)
  plt.show()

metric = SPDMetricBuresWasserstein(2)
dis = metric.squared_dist(samples[0] , samples[1])
print(dis)
```
